```python
import numpy as np 
import tensorflow as tf 
from tensorflow.python.tools.inspect_checkpoint import  print_tensors_in_checkpoint_file
# tf.reset_default_graph函数用于清除默认图形堆栈并重置全局默认图形。 放在前面一点。
tf.reset_default_graph()
saver=tf.train.import_meta_graph('./mnist_model/mnist_model.ckpt.meta')
graph=tf.get_default_graph() 
graph
```

### 得到当前图中所有变量的名称

```python
tensor_name_list = [tensor.name for tensor in graph.as_graph_def().node]# 得到当前图中所有变量的名称
print(tensor_name_list)
```

### 获取测试数据


```python
from tensorflow.examples.tutorials.mnist import input_data 
mnist=input_data.read_data_sets('./fashion-mnist/')
X,y=mnist.test.images,mnist.test.labels
```

### 加载模型

```python
sess=tf.Session() 
saver.restore(sess,'./mnist_model/mnist_model.ckpt')
x_i=graph.get_tensor_by_name("input_x:0") # x作为PlaceHolder时的name,要加 0 
y_o=graph.get_tensor_by_name('output_y:0') # y作为y_hat 时的name,要加 0 
```

### 预测实验 

```python
t=tf.argmax(y_o,1)  
# f=np.expand_dims(mnist.test.images[0],0) ，预测单个时需要扩1个维度
y_hat=t.eval(session=sess,feed_dict={x_i:mnist.test.images})  
y_hat
```

```python
(y_hat==y ).mean() 
```
