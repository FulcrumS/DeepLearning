# 获取数据

```python
import mxnet.ndarray as nd
import mxnet.autograd as ag
features = 2
examples = 100
X = nd.random_normal(shape=(examples, features))
true_w = [1.4, 0.4]
true_b = 1.2
Y = true_w[0]*X[:, 0]+true_w[1]*X[:, 1]+true_b
Y += 0.1*nd.random_normal(shape=Y.shape)
X, Y
```

# 定义训练的权重和b
## w、b都要向系统获得求导的空间

```python
w = nd.random_normal(shape=(features, 1))
b = nd.zeros((1,))
params = [w, b]
for param in params:
    param.attach_grad()
```

# 计算训练的X

```python
def net(X):
    return nd.dot(X, w)+b
```

# 定义误差函数

```python
def square_loss(yHat, y):
#     print("***",(yHat-y.reshape(yHat.shape))**2)
#     print('***',yHat)
    return (yHat-y.reshape(yHat.shape))**2
```

# 定义优化函数
## 就地修改：w、b

```python
def SGD(params, lr):
    for param in params:
        param[:] = param-lr*param.grad
```

# 随机获得batch_size个训练样本和标签

```python
batch_size = 10


def data_iter():
    idx = list(range(examples))
    import random
    random.shuffle(idx)
    for i in range(0, examples, batch_size):
        j = nd.array(idx[i:min(i+batch_size, examples)])
        yield nd.take(X, j), nd.take(Y, j)
```

# 定义学习率、在epochs轮完成误差计算和w、b的更新

```python
epochs = 6
learning_rate = 0.01
for e in range(epochs):
    total_loss = 0
    for data, label in data_iter():
        with ag.record():
            output = net(data)
            loss = square_loss(output, label)
        loss.backward()
        SGD(params, learning_rate)
        total_loss += nd.sum(loss).asscalar()
    print("Epoch:%d ,avg loss:%f " % (e+1, total_loss/examples))
```
