```python
from mxnet import gluon
from mxnet import nd
from mxnet import autograd


def transform(data, label):
    return data.astype('float32')/255, label.astype('float32')


minist_train = gluon.data.vision.FashionMNIST(
    root='fashion-mnist/', train=True, transform=transform)
minist_test = gluon.data.vision.FashionMNIST(
    root='fashion-mnist/', train=False, transform=transform)
```

```python
def show_image(images):
    import matplotlib.pyplot as plt
    n = images.shape[0]
    _, fig = plt.subplots(1, n, figsize=(15, 15))
    for i in range(n):
        fig[i].imshow(images[i].reshape(28, 28).asnumpy())
        fig[i].axes.get_xaxis().set_visible(False)
        fig[i].axes.get_yaxis().set_visible(False)
    plt.show()


def get_text_labels(labels):
    class_names = ['短袖圆领T恤', '裤子', '套衫', '连衣裙', '外套',

                   '凉鞋', '衬衫', '运动鞋', '包', '短靴']
    return [class_names[int(i)] for i in labels]


# data, label = minist_train[0:9]
# show_image(data)
# get_text_labels(label)
from utils import load_data_fashion_mnist as load 
batch_size=5
train_iter,test_iter=load.load_data_fashion_mnist(batch_size)
for X,y in train_iter:
    show_image(X)
    print(X.shape,y)
    break;
```

```python
batch_size=256 
train_data=gluon.data.DataLoader(minist_train,batch_size=batch_size,shuffle=True)
test_data=gluon.data.DataLoader(minist_test,batch_size=batch_size,shuffle=True)
```

```python
num_inputs=28*28 
num_outputs=10 
w=nd.random_normal(shape=(num_inputs,num_outputs))
b=nd.random_normal(shape=(num_outputs))
params=[w,b]
params
```

```python
for param in params:
    param.attach_grad()
```

```python
from mxnet import nd


def softmax(x):
    exp = nd.exp(x)
    partition = exp.sum(axis=1, keepdims=True)
    return exp/partition
```

```python
# x=nd.random_normal(shape=(2,4))
# prob=softmax(x)
# # x,prob,prob.sum(axis=1 )
# y=nd.array([0,1,0,0]).reshape((4,1 ))
# nd.pick(x,y)
```

```python
def SGD(params,lr):
    for param in params:
        param[:]=param-lr*param.grad
```

```python
def net(X):
    return softmax(nd.dot(X.reshape((-1,num_inputs)),w)+b)
```

```python
er=0
def cross_entropy(yhat,y):
    global er 
    if er==0: 
        print('-----')
        print(nd.log(yhat).shape,y.shape) 
        print(nd.log(yhat)[2],y[2])
        print('****\n',-nd.pick(nd.log(yhat),y)[2])
        print('-----')
        er+=1
    return -nd.pick(nd.log(yhat),y)
```

```python
from mxnet import nd 
nd.one_hot(nd.array([1,2,1]))
```

```python
def accuracy(output,label):
    return nd.mean(output.argmax(axis=1)==label).asscalar()
```

```python
def evaluate_accuracy(data_iter,net):
    acc=0. 
    for data,label in data_iter:
        output=net(data )
        acc+=accuracy(output,label)
    return acc/len(data_iter)
```

```python
# evaluate_accuracy(train_data,net)
```

```python
learning_rate=.1 
epochs=5
for e in range(1,epochs+1 ):
    train_loss,train_acc=0.,0. 
    for data,label in train_data:
        with autograd.record():
            output=net(data)
            loss=cross_entropy(output,label)
        loss.backward()
        SGD(params,learning_rate/batch_size)
        train_loss+=nd.mean(loss).asscalar()
        train_acc+=accuracy(output,label)
    test_acc=evaluate_accuracy(test_data,net)
    print('epoch %d,loss:%f..,train_accuracy:%f,test_accuracy:%f..'%(e,train_loss/len(train_data),train_acc/len(train_data),
                                                                    test_acc/len(test_data)))
        
```
